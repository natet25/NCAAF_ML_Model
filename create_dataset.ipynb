```python
import re
def get_school_name(x):
    school = re.findall("[-A-Za-z ]+[A-Za-z()&]* ", x)
    return school[0].rstrip()
```

```python
rankings_dfs = []
for year in range(2011, 2022):
    filename = 'ap_rankings_' + str(year) + '.xlsx'
    rankings = pd.read_excel(filename)
    rankings['year'] = year
    rankings = rankings[rankings['Date'] == 'Final']
    rankings['school_name'] = rankings['School'].apply(get_school_name)
    rankings = rankings[['Rk', 'school_name', 'year']]
    rankings_dfs.append(rankings)
ap_rankings = pd.concat(rankings_dfs)
```

```python
data = cfb.merge(ap_rankings, how='left', left_on=['name', 'year'], right_on = ['school_name', 'year'])
data.columns
```

```python
data = data.drop(['ranking', 'school_name'], axis=1)
data.to_csv('CFBData.csv')
```
